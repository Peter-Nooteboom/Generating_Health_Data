```python
from google.colab import drive
drive.mount('/content/drive')
directory = '/content/drive/MyDrive/Data Projects/Data Generation/Files'

import pandas as pd
```

```python
df = pd.read_csv(f'{directory}/Original_Data.csv')
```

```python
df.head(1)
```

```python
#Reordering and Renaming
df = df[['SOURCE','SEX','AGE', 'HAEMATOCRIT',	'HAEMOGLOBINS',	'ERYTHROCYTE',	'LEUCOCYTE',	'THROMBOCYTE',	'MCH',	'MCHC',	'MCV']]
df = df.rename(columns={'SOURCE':'Next_Treatment_Type', 'SEX':'Sex', 'AGE':'Age', 'HAEMATOCRIT':'Hematocrit', 'HAEMOGLOBINS':'Hemoglobins', 'ERYTHROCYTE':'Erythrocyte', 'LEUCOCYTE':'Leucocyte', 'THROMBOCYTE':'Thrombocyte'})
```

```python
df.head(1)
```

```python
#Recoding Variables
df['Next_Treatment_Type'].replace('out', 0, inplace=True)
df['Next_Treatment_Type'].replace('in', 1, inplace=True)
df['Sex'].replace('M', 0, inplace=True)
df['Sex'].replace('F', 1, inplace=True)
```

```python
df.head(5)
```

```python
#Saving Data File
df.to_csv(f'{directory}/EHR_Data.csv', index=False)
```
